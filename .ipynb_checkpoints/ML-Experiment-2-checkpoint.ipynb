```python
titanic.drop(['Cabin'], axis=1, inplace=True)
```

```python
titanic.dropna(inplace=True)
```

```python
titanic
```

```python
titanic.isnull().sum()
```

```python
new_sex=pd.get_dummies(titanic['Sex'], drop_first=True)
```

```python
new_pclass=pd.get_dummies(titanic['Pclass'], drop_first=True)
```

```python
new_embarked=pd.get_dummies(titanic['Embarked'], drop_first=True)
```

```python
titanic=pd.concat([titanic,new_sex,new_pclass,new_embarked],axis=1)
```

```python
titanic
```

```python
titanic.drop(['PassengerId','Pclass','Name','Ticket','Sex','Embarked'],axis=1,inplace=True)
```

```python
titanic
```

```python
y=titanic['Survived']
```

```python
titanic.drop(['Survived'],axis=1,inplace=True)
```

```python
x=titanic
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg =  LogisticRegression()
```

```python
x_train.columns = x_train.columns.astype(str)
x_test.columns = x_test.columns.astype(str)
logreg.fit(x_train,y_train)
```

```python
prediction = logreg.predict(x_test) 
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test,prediction)
```
