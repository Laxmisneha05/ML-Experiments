# Implement SVM with Bank Loan dataset for multiclass classification and evalute it using various performance measures

```python
import pandas as pd
import numpy as np
from sklearn import svm, datasets
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('bank-loan.csv')
df.head()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df['ed'].value_counts().plot(kind='bar')
```

```python
sns.pairplot(data=df, hue='ed')
```

```python
df
```

```python
X = df
y = df['default']
X.drop(['default'],axis=1,inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y)
encoded_y = le.transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, encoded_y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='poly')

# fitting x samples and y classes
model.fit(X_train, y_train)
```

```python
def prediction(a):
  global model,le
  pred = model.predict(a)
  ans = le.inverse_transform(pred)
  return ans
```

```python
y_pred = prediction(X_test)
```

```python
y_actual = le.inverse_transform(y_test)
```

```python
from sklearn import metrics
accuracy=metrics.accuracy_score(y_pred, y_actual)*100
print('Accuracy of our model is equal ' + str(round(accuracy, 2)) + ' %.')
```
