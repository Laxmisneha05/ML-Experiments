# Implement Agglomerative clustering using a E-Commerce Customers dataset.

```python
import numpy as np
import pandas as pd
dataset = pd.read_csv('Ecommerce Customers.csv')
x = dataset.iloc[:, [3,4]].values
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h = AgglomerativeClustering(n_clusters = 5 , affinity='euclidean', linkage='ward')
y = h.fit_predict(x)
y
```

```python
plt.scatter(x[y == 0,0], x[y == 0,1], s=100, c='red', label = 'Carefull')
plt.scatter(x[y == 1,0], x[y == 1,1], s=100, c='blue', label = 'General')
plt.scatter(x[y == 2,0], x[y == 2,1], s=100, c='green', label = 'Target')
plt.scatter(x[y == 3,0], x[y == 3,1], s=100, c='cyan', label = 'Careless')
plt.scatter(x[y == 4,0], x[y == 4,1], s=100, c='grey', label = 'Miser')
plt.title('Clusters of customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
